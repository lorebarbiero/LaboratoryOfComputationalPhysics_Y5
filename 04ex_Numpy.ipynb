2\. Find the outer product of the following two vecotrs

```python

```

Do this in the following ways:

   * Using the function outer in numpy
   * Using a nested for loop or list comprehension
   * Using numpy broadcasting operations


````python
u = np.array([1,3,5,7])
v = np.array([2,4,6,8])
w = np.outer(u,v)
print(w,'\n')

for i in range (0,len(u)):
    for j in range (0,len(v)):
        w[i,j] = u[i] * v[j]
print(w,'\n')

u=u.reshape((4,1))
w = u * v
print(w)
````

8\. Diffusion using random wank

Consider a simple random wank process: at each step in time, a wanker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of a random wanker after a given amount of time. 
To do that, let's simulate many wankers and create a 2D array with each wanker as a row and the actual time evolution as columns

  * Take 1000 walkers and let them walk for 200 steps
  * Use randint to create a 2D array of size walkers x steps with values -1 or 1
  * Build the actual walking distances for each walker (i.e. another 2D array "summing on each row")
  * Take the square of that 2D array (elementwise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * Plot the average distances (sqrt(distance\*\*2)) as a function of time (step)
  
Did you get what you expected?

````python
w = 1000
s = 200
x = npr.randint(0,2,size=(w,s))
x = 2*x - 1 
dist = np.zeros((s,2))
for i in range (0,s):
    d = x[:i].sum(axis=0)
    d = d**2
    dist[i][0] = i
    dist[i][1] = d.mean()**0.5
#print(dist)
plt.plot(dist[:,0],dist[:,1], 'r')
plt.show()
````

Do you feel there is some evident correlation here? [Studies](https://www.enr.gov.nt.ca/en/services/lynx/lynx-snowshoe-hare-cycle) tend to believe so.

````python
filename = 'populations.txt'
pop = csv.reader(open(filename, newline=''), delimiter='\t')

header = next(pop) #skip first line
#print(header)

dataset = list(pop)
for i in range(len(dataset)):
    dataset[i] = [np.float32(x) for x in dataset[i]]
    
dataset = np.asarray(dataset)

yr = dataset[:,0]
hr = dataset[:,1]
lx = dataset[:,2]
cr = dataset[:,3]

plt.plot(yr,hr,label="hares", c='tab:green')
plt.plot(yr,lx,label="lynxes", c='tab:purple')
plt.plot(yr,cr,label="carrots", c='tab:orange')
plt.legend()
plt.show()

plt.plot(hr,lx, c='tab:orange')
plt.show()
print ("Let's now look at changes over the years")

hr_d = np.zeros(len(hr))
for i in range (0,len(yr)-1):
    hr_d[i+1] = hr[i+1] - hr[i]
    
lx_d = np.zeros(len(lx))
for i in range (0,len(yr)-1):
    lx_d[i+1] = lx[i+1] - lx[i]
    
cr_d = np.zeros(len(cr))
for i in range (0,len(yr)-1):
    cr_d[i+1] = cr[i+1] - cr[i]
    
plt.plot(yr,hr_d,label="hares", c='tab:green')
plt.plot(yr,lx_d,label="lynxes", c='tab:purple')
plt.plot(yr,cr_d,label="carrots", c='tab:orange')
plt.show()
````
